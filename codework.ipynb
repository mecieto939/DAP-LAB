```python
#Question no 1
n = int(input("Enter an integer: "))
total = 0

while n > 0:
    total += n % 10
    n //= 10

print("Sum of digits:", total)
```

```python
#Question no 2
n = int(input("Enter an integer: "))
product = 1

while n > 0:
    product *= n % 10
    n //= 10

print("Product of digits:", product)
```

```python
#Question no 3
n = int(input("Enter a number: "))
reverse = 0

while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n //= 10

print("Reversed number:", reverse)
```

```python
#Question no 4
n = int(input("Enter a number: "))
temp = n
reverse = 0

while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n //= 10

if temp == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

```python
#Question no 5
n = int(input("Enter a number: "))
temp = n


digits = 0
t = n
while t > 0:
    digits += 1
    t //= 10


armstrong_sum = 0
t = n
while t > 0:
    digit = t % 10
    armstrong_sum += digit ** digits
    t //= 10


if armstrong_sum == temp:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```
